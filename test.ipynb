```python
class Solution:
    def solve(self,n):
        num=1
        top=0
        bottom=n-1
        left=0
        right=n-1
        mat=[[0]*n for _ in range(n)]

        for i in range(n):

            for j in range(left,right+1):
                    mat[top][j]=num
                    num+=1
            top+=1
            for j in range(top,bottom+1):
                mat[j][right]=num
                num+=1

            right-=1
            for j in range(right,left-1,-1):
                mat[bottom][j]=num
                num+=1
            
            bottom-=1
            for j in range(bottom,top-1,-1):
                mat[j][left]=num
                num+=1
            
            left+=1
           
        return mat

obj=Solution()
mat=obj.solve(5)
for i in range(5):
    print(mat[i])
```
