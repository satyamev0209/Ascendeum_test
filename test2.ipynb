```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("analysis_id.csv")
df.head(10)
```

```python
df.columns
```

```python
features=["bidder", "response_count","prebid_win_count"]
df=df[features]
df.head(10)
```

```python
df.isna().sum()
```

```python
df["prebid_win_count"]=df["prebid_win_count"].fillna(0)
df.isna().sum()
```

```python
summary = (df.groupby("bidder").agg(wins=("prebid_win_count","sum"), total_bid=("response_count","sum"))).reset_index()
summary["win_rate"]=round(((summary["wins"])/(summary["total_bid"])*100),3).sort_values(ascending=False)
```

```python
summary=summary[1:]
summary=summary.sort_values(["win_rate"], ascending=False)
```
